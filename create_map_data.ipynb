```python
total_weather = pd.read_csv('C:/Users/20175878/Documents/GitHub/2AMD20-group19/2_preprocessed_data/TotaalWeerStation.csv')
weerstation_per_gemeente = pd.read_csv('C:/Users/20175878/Documents/GitHub/2AMD20-group19/2_preprocessed_data/weerstation_per_gemeente.csv')

joined = joined.merge(weerstation_per_gemeente, on='Gemeentenaam', how='left')
joined = joined.merge(total_weather, on='WeerStation', how='left')
joined
```

```python
inhabitants = pd.read_csv('C:/Users/20175878/Documents/GitHub/2AMD20-group19/2_preprocessed_data/InhabitantsMunicipality.csv')
joined = joined.merge(inhabitants, on='Gemeentenaam', how='left')
joined
```

```python
postcodes = pd.read_csv('C:/Users/20175878/Documents/GitHub/2AMD20-group19/2_preprocessed_data/postcodeMunicipality.csv')
restricties = pd.read_csv('C:/Users/20175878/Documents/GitHub/2AMD20-group19/2_preprocessed_data/Restrictions.csv')

pr = postcodes.merge(restricties, on='Postcode', how='left')
pr_grouped = pr.groupby('Gemeentecode')['Kleur'].agg(list)

def remove_nans(mylist):
    cleaned_list = [x for x in mylist if str(x) != 'nan']

    if len(cleaned_list) == 0:
        return 'NO DATA'
    else:
        return max(set(cleaned_list), key = cleaned_list.count)

pr_grouped = pr_grouped.apply(remove_nans).reset_index()

pr_grouped
joined = joined.merge(pr_grouped, on='Gemeentecode', how='left')
joined
```

```python
sun = pd.read_csv('C:/Users/20175878/Documents/GitHub/2AMD20-group19/2_preprocessed_data/sun.csv')
joined = joined.merge(sun, on='Gemeentenaam', how='left')
joined
```
