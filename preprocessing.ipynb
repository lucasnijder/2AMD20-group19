```python
dfTotaalWeerStations = pd.DataFrame()

path = "1_source_data/klimaattabellen"
WS_list = os.listdir(path)

for WS in WS_list:
    urlWeerStation = f"https://raw.githubusercontent.com/lucasnijder/2AMD20-group19/main/1_source_data/klimaattabellen/{WS}"
    ColumnNames = ['metric', 'jan', 'feb', 'mar', 'apr', 'may', 'jun', 'jul', 'aug', 'sep', 'okt', 'nov', 'dec', 'winter', 'lente', 'zomer', 'herfst', 'jaar']
    dfWeerStation = pd.read_csv(urlWeerStation, encoding= "unicode_escape", sep = ",", comment='#', names=ColumnNames).set_index('metric')['jaar']
    dfTotaalWeerStations[WS[:-4]] = dfWeerStation
    
dfTotaalWeerStations = dfTotaalWeerStations.T
```

```python
dfTotaalWeerStations.columns = dfTotaalWeerStations.columns.map(lambda x: x.strip(" "))

columnsNecessary_WeerStation = [col for col in dfTotaalWeerStations.columns if col.startswith('FXX') or col.startswith('pct') or col.startswith('FG')]
columnsNecessary_WeerStation.append('TG_gemiddeld')

preProcessed_TotaalWeerStation = dfTotaalWeerStations[columnsNecessary_WeerStation].reset_index()
preProcessed_TotaalWeerStation = preProcessed_TotaalWeerStation.rename_axis(None, axis=1)
preProcessed_TotaalWeerStation = preProcessed_TotaalWeerStation.rename({'index':'WeerStation'}, axis=1)
preProcessed_TotaalWeerStation['WeerStation'] = preProcessed_TotaalWeerStation['WeerStation'].str.replace('_', ' ').str.upper()

preProcessed_TotaalWeerStation = preProcessed_TotaalWeerStation[['WeerStation','pctzon_gemiddeld','FG_gemiddeld']]

preProcessed_TotaalWeerStation.to_csv('2_preprocessed_data/TotaalWeerStation.csv', index=False)
preProcessed_TotaalWeerStation
```

```python
dfLocatie_Weerstations = pd.DataFrame(columns=["LON(east)", "LAT(north)", "ALT(m)", "WeerStation"])
path = "1_source_data/klimaattabellen"
WS_list = os.listdir(path)


for WS in WS_list:
    #hoek van holland is in another format for some reason
    if WS != "Hoek_van_Holland.csv":
        urlWeerStation = f"https://raw.githubusercontent.com/lucasnijder/2AMD20-group19/main/1_source_data/klimaattabellen/{WS}"
        # we get just the ninth row, pandas will interpret it as a dataframe with 1 column and no rows.
        dfWeerStation = pd.read_csv(urlWeerStation, encoding= "unicode_escape", sep = "," ,skiprows=lambda x: x != 9) 
        # so we take the name of that column as our raw lat long string
        # its in the format:
        #"# STN:         LON(east)       LAT(north)      ALT(m)  NAME"
        raw = dfWeerStation.columns[0]
        # format it into an array of format [LON(east), LAT(norht), ALT(m), NAME]
        raw = raw.split("  ")[1:]
        raw[:] = [x.strip() for x in raw if x]
        
        #turn it into a dataframe and concat it with the previous entries
        data = pd.DataFrame(pd.Series(raw, index=["LON(east)", "LAT(north)","ALT(m)","WeerStation"])).T
        dfLocatie_Weerstations = pd.concat([dfLocatie_Weerstations, data], axis=0, ignore_index=True)

dfLocatie_Weerstations['WeerStation'] = dfLocatie_Weerstations['WeerStation'].str.strip()

dfLocatie_Weerstations.to_csv('2_preprocessed_data/LocatieWeerStation.csv', index=False)
dfLocatie_Weerstations
```

```python
urlgeolocMunicipality = "https://raw.githubusercontent.com/lucasnijder/2AMD20-group19/main/1_source_data/Alle-gemeenten-in-Nederland.xls"
dfgeolocMunicipality = pd.read_excel(urlgeolocMunicipality) #, header=0, encoding= "unicode_escape", sep =";" )

lattitudes = []
longitudes = []
for i, row in dfgeolocMunicipality.iterrows():
        rawcoords = row["geometry"].split(",")
        processedcoords = []
        #idk why, but some coordinates start with "0.0" before the actual coordinate
        #Its probably alright to remove those
        for coord in rawcoords:
            coord = ''.join(c for c in coord if (c.isdigit() or c =='.'))
            if coord[:3] == "0.0":
                coord = coord[3:]
            if coord != "":
                processedcoords.append(float(coord))
        
        
        tot_lat = 0
        tot_lon = 0
        
        if len(processedcoords) % 2 != 0:
            print("oh noes something is wrong")
            
        #add together all lattitude and longitude values
        for i in range(len(processedcoords)):
            if i%2 == 0:
                tot_lon += processedcoords[i]
            else:
                tot_lat += processedcoords[i]
        
        avg_lon = tot_lon/len(processedcoords) *2        
        avg_lat = tot_lat/len(processedcoords) *2
        
        
        longitudes.append(avg_lon)
        lattitudes.append(avg_lat)
    
dfgeolocMunicipality["LAT(north)"] = lattitudes        
dfgeolocMunicipality["LON(east)"] = longitudes

dfgeolocMunicipality.drop('GM_CODE', inplace=True, axis=1)
dfgeolocMunicipality.drop('geometry', inplace=True, axis=1)

dfgeolocMunicipality.rename(columns={'GM_NAAM': 'Gemeentenaam'}, inplace=True)

dfgeolocMunicipality.to_csv('2_preprocessed_data/geolocMunicipality.csv', index=False)
dfgeolocMunicipality
```

```python
# aftstanden

import geopy.distance

url_gem_locs = "https://raw.githubusercontent.com/lucasnijder/2AMD20-group19/main/2_preprocessed_data/geolocMunicipality.csv"
df_gem_locs = pd.read_csv(url_gem_locs).dropna() #, header=0, encoding= "unicode_escape", sep =";" )
url_weer_locs = "https://raw.githubusercontent.com/lucasnijder/2AMD20-group19/main/2_preprocessed_data/LocatieWeerStation.csv"
df_weer_locs= pd.read_csv(url_weer_locs).dropna()

df_gem_locs[['LAT(north)','LON(east)']] = df_gem_locs[['LAT(north)','LON(east)']].astype('float')
df_weer_locs[['LAT(north)','LON(east)']] = df_weer_locs[['LAT(north)','LON(east)']].astype('float')

df_distances = pd.DataFrame(columns=["Gemeentenaam", "distance", "WeerStation"])

for gem in df_gem_locs.iterrows():
    
    for weer in df_weer_locs.iterrows():
        gem_lon = gem[1].get("LON(east)")
        gem_lat = gem[1].get("LAT(north)")
        
        gem_coords = (gem_lat, gem_lon)
        
        weer_lon = weer[1].get("LON(east)")
        weer_lat = weer[1].get("LAT(north)")
        
        weer_coords = (weer_lat, weer_lon)
        
        distance =  geopy.distance.geodesic(gem_coords, weer_coords).km
        
        entry=[gem[1].get("Gemeentenaam"), distance, weer[1].get("WeerStation")]
    #     print(entry)
        #turn it into a dataframe and concat it with the previous entries
        data = pd.DataFrame(pd.Series(entry, index=["Gemeentenaam", "distance", "WeerStation"])).T
        df_distances = pd.concat([df_distances, data], axis=0, ignore_index=True)

df_distances.to_csv('2_preprocessed_data/gem_weer_distances.csv', index=False)
df_distances
```
