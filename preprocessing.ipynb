WINDENERGIE

varname = Wind

```python
urlWind = "https://raw.githubusercontent.com/lucasnijder/2AMD20-group19/main/1_source_data/Windenergie_op_land__per_provincie_24052022_142754.csv"
dfWind = pd.read_csv(urlWind, header=0, encoding= "unicode_escape", sep =";" )

columnRenaming_wind = {"ï»¿\"Regio's\"":'Regio', 
                        "Perioden":"Periode"}
columnsNecessary_wind = ['Regio','Periode',
                            'Elektriciteitsproductie/Genormaliseerde productie (mln kWh)',
                        	'Elektriciteitsproductie/Productiefactor (%)',
                            'Elektrisch vermogen/Opgesteld vermogen einde jaar (megawatt)']
                            
preProcessed_dfWind = dfWind.rename(columnRenaming_wind, axis=1)
preProcessed_dfWind = preProcessed_dfWind[columnsNecessary_wind]
preProcessed_dfWind = preProcessed_dfWind[preProcessed_dfWind['Periode'] == 2019]
preProcessed_dfWind = preProcessed_dfWind.drop('Periode', axis=1)

preProcessed_dfWind.to_csv('2_preprocessed_data/Wind.csv')
preProcessed_dfWind
```

ENERGIEVERBRUIK PARTICULIER

varname = energyUsage

<sup>--> vraag: willen we iets doen met woningkenmerken (e.g. rijtjeshuis)?<sup>

```python
urlEnergyUsage = "https://raw.githubusercontent.com/lucasnijder/2AMD20-group19/main/1_source_data/energieverbruik%20particuliere%20woniningen%20-%2084837NED_UntypedDataSet_21052022_135719.csv"
dfEnergyUsage = pd.read_csv(urlEnergyUsage, header=0, encoding= "unicode_escape", sep =";")

columnRenaming_energyUsage = {"Gemeentenaam_1":'Gemeentenaam', 
                        "GemiddeldAardgasverbruik_4":"GemiddeldAardgasverbruik",
                        "GemiddeldElektriciteitsverbruik_5":"GemiddeldElektriciteitsverbruik"}
columnsNecessary_energyUsage = ["Gemeentenaam", "GemiddeldAardgasverbruik","GemiddeldElektriciteitsverbruik"]

preProcessed_dfEnergyUsage = dfEnergyUsage.rename(columnRenaming_energyUsage, axis=1)
preProcessed_dfEnergyUsage = preProcessed_dfEnergyUsage[columnsNecessary_energyUsage]

preProcessed_dfEnergyUsage["GemiddeldAardgasverbruik"] = pd.to_numeric(preProcessed_dfEnergyUsage["GemiddeldAardgasverbruik"], errors="coerce")
preProcessed_dfEnergyUsage["GemiddeldElektriciteitsverbruik"] = pd.to_numeric(preProcessed_dfEnergyUsage["GemiddeldElektriciteitsverbruik"], errors="coerce")

grouped_preProcessed_dfEnergyUsage = preProcessed_dfEnergyUsage.groupby(["Gemeentenaam"])[["GemiddeldAardgasverbruik","GemiddeldElektriciteitsverbruik"]].sum().reset_index()

grouped_preProcessed_dfEnergyUsage.to_csv('2_preprocessed_data/EnergyUsage.csv')
grouped_preProcessed_dfEnergyUsage
```

INWONERS PER GEMEENTE

varname = InhabitantsMunicipality

```python
urlInhabitantsMunicipality = "https://raw.githubusercontent.com/lucasnijder/2AMD20-group19/main/1_source_data/inwoners%20per%20gemeente.csv"
dfInhabitantsMunicipality = pd.read_csv(urlInhabitantsMunicipality, header=0, encoding= "unicode_escape", sep =";" )

columnRenaming_inhabitantsMunicipality = {'Gemeente':'Gemeentenaam'}
columnsNecessary_inhabitantsMunicipality = ["Gemeentenaam", "Inwoneraantal" ,"Groot, middelgroot en klein"]
                            
preProcessed_inhabitantsMunicipality = dfInhabitantsMunicipality.rename(columnRenaming_inhabitantsMunicipality, axis=1)
preProcessed_inhabitantsMunicipality = preProcessed_inhabitantsMunicipality[columnsNecessary_inhabitantsMunicipality]

preProcessed_inhabitantsMunicipality["Inwoneraantal"] = pd.to_numeric(preProcessed_inhabitantsMunicipality["Inwoneraantal"], errors="coerce")
preProcessed_inhabitantsMunicipality = preProcessed_inhabitantsMunicipality[preProcessed_inhabitantsMunicipality["Inwoneraantal"] > 30000]
preProcessed_inhabitantsMunicipality['Gemeentenaam'] = preProcessed_inhabitantsMunicipality['Gemeentenaam'].map(lambda x: x.strip("?"))

preProcessed_inhabitantsMunicipality.to_csv('2_preprocessed_data/InhabitantsMunicipality.csv')
preProcessed_inhabitantsMunicipality
```

ZONNESTROOM VERMOGEN

varname = sun

```python
urlSun = "https://raw.githubusercontent.com/lucasnijder/2AMD20-group19/main/1_source_data/zonnestroom%20vermogen%202019%20-%2085010NED_UntypedDataSet_21052022_134211.csv"
dfSun = pd.read_csv(urlSun, header=0, encoding= "unicode_escape", sep =";" )

columnRenaming_sun = {"Gemeentenaam_1":"Gemeentenaam",
                    "AantalInstallatiesBijWoningen_5":"AantalInstallatiesBijWoningen",
                    "OpgesteldVermogenVanZonnepanelen_6":"OpgesteldVermogenVanZonnepanelen"}
columnsNecessary_sun = ["Gemeentenaam", "AantalInstallatiesBijWoningen" , "OpgesteldVermogenVanZonnepanelen"]
                            
preProcessed_sun = dfSun.rename(columnRenaming_sun, axis=1)
preProcessed_sun = preProcessed_sun[columnsNecessary_sun]

grouped_preProcessed_sun = preProcessed_sun.groupby(["Gemeentenaam"])[["AantalInstallatiesBijWoningen", "OpgesteldVermogenVanZonnepanelen"]].sum().reset_index()
grouped_preProcessed_sun = grouped_preProcessed_sun[grouped_preProcessed_sun['Gemeentenaam'] != '                                        ']
grouped_preProcessed_sun['Gemeentenaam'] = grouped_preProcessed_sun['Gemeentenaam'].map(lambda x: x.strip(" "))

grouped_preProcessed_sun.to_csv('2_preprocessed_data/Sun.csv')
grouped_preProcessed_sun
```
